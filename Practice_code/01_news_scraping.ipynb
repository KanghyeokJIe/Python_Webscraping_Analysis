- ì›¹ ìŠ¤í¬ë˜í•‘ ì‹œ **ì ì ˆí•œ ì¸ì½”ë”© ì„¤ì •** (`res.encoding`)
- CSS ì„ íƒìë¥¼ ì´ìš©í•œ **BeautifulSoupì˜ select() ì‚¬ìš©ë²•**

- ì¶œë ¥ ìˆœì„œì™€ í¬ë§·ì„ **ë¬¸ì œ ì¡°ê±´ì— ë§ê²Œ ì •ë¦¬í•˜ì—¬ êµ¬í˜„**

```python
import requests     
from bs4 import BeautifulSoup

# 1. ìš”ì²­ í—¤ë” ì„¤ì • (ë¸Œë¼ìš°ì €ì²˜ëŸ¼ ë³´ì´ê²Œ í•˜ê¸° ìœ„í•œ User-Agent ì„¤ì •)
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'
}

# 2. URLì— ë„£ì„ íŒŒë¼ë¯¸í„° (sid: ì„¹ì…˜ ì•„ì´ë”” - economy = ê²½ì œ ì„¹ì…˜)
req_param = {
    'sid': 'economy'
}

# 3. ì‹¤ì œ ìš”ì²­í•  URL ìƒì„±
url = 'https://news.daum.net/{sid}'.format(**req_param)
print(url)  # â†’ https://news.daum.net/economy

# 4. ì›¹ ì„œë²„ì— ìš”ì²­ (í—¤ë” í¬í•¨)
res = requests.get(url, headers=req_header)

# 5. ì¸ì½”ë”© ì„¤ì • (ì¤‘ìš”! í•œê¸€ ê¹¨ì§ ë°©ì§€)
res.encoding = 'utf-8'

# 6. ì‘ë‹µ ìƒíƒœ ë° íƒ€ì… í™•ì¸
print(type(res))           # <class 'requests.models.Response'>
print(res.status_code)     # 200ì´ë©´ ì •ìƒ

# 7. ì‘ë‹µì´ ì •ìƒì¼ ë•Œë§Œ íŒŒì‹± ì§„í–‰
if res.ok:
    html = res.text  # HTML ì½”ë“œ ë¬¸ìì—´
    soup = BeautifulSoup(html, 'html.parser')  # BeautifulSoupë¡œ HTML íŒŒì‹±

    # 8. ë‰´ìŠ¤ ê¸°ì‚¬ ë§í¬ê°€ í¬í•¨ëœ a íƒœê·¸ ë¦¬ìŠ¤íŠ¸ ì¶”ì¶œ (hrefì— '/v.daum.net' í¬í•¨)
    li_tag_list = soup.select("ul.list_newsheadline2 a[href*='/v.daum.net']")
    print(type(li_tag_list), len(li_tag_list))  # ëª‡ ê°œ ìˆëŠ”ì§€ ì¶œë ¥

    # 9. ê°ê°ì˜ ë‰´ìŠ¤ í•­ëª©(li íƒœê·¸) ìˆœíšŒí•˜ë©° ì œëª©ê³¼ ë§í¬ ì¶œë ¥
    for li_tag in soup.select('ul.list_newsheadline2 li'):
        a_tag = li_tag.find('a')  # ë‰´ìŠ¤ ê¸°ì‚¬ë¡œ ê°€ëŠ” ë§í¬
        link = a_tag['href']      # href ì†ì„±ê°’ (URL)
        print(link)

        # ë‰´ìŠ¤ ì œëª©ì´ ë‹´ê¸´ strong íƒœê·¸ ì„ íƒ
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text.strip()  # í…ìŠ¤íŠ¸ ì¶”ì¶œ í›„ ê³µë°± ì œê±°
        print(title)
```

## ğŸ’¡ í•™ìŠµ í¬ì¸íŠ¸

- `ë”•ì…”ë„ˆë¦¬ ë§¤í•‘`ì„ í™œìš©í•œ ì…ë ¥ ì²˜ë¦¬ ë° ìœ ì—°í•œ URL êµ¬ì„±
- `í•¨ìˆ˜í™”`ë¥¼ í†µí•œ ì½”ë“œ ì¬ì‚¬ìš©ì„±ê³¼ í™•ì¥ì„± í™•ë³´
- `BeautifulSoup`ì„ ì‚¬ìš©í•œ `select()`, `select_one()`ë¡œ ì›í•˜ëŠ” íƒœê·¸ ì¶”ì¶œ
- ì›¹ í˜ì´ì§€ í¬ë¡¤ë§ ì‹œ **ì¸ì½”ë”© ì„¤ì •** ì¤‘ìš”ì„± ì´í•´



```python
import requests
from bs4 import BeautifulSoup

# í•œê¸€ â†’ ì˜ì–´ ì„¹ì…˜ ì½”ë“œ ë”•ì…”ë„ˆë¦¬
# ë‰´ìŠ¤ URLì—ëŠ” ì˜ì–´ ì½”ë“œê°€ ë“¤ì–´ê°€ê³ , ì¶œë ¥ì—ëŠ” í•œê¸€ì„ ê·¸ëŒ€ë¡œ ì“°ê¸° ìœ„í•´ ì‚¬ìš©
section_dict = {
    'ìµœì‹ ë‰´ìŠ¤': 'climate',
    'ì‚¬íšŒ': 'society',
    'ê²½ì œ': 'economy',
    'ì •ì¹˜': 'politics',
    'êµ­ì œ': 'world',
    'ë¬¸í™”': 'culture',
    'ìƒí™œ': 'life',
    'IT/ê³¼í•™': 'tech',
    'ì¸ë¬¼': 'people'
}

# ë‰´ìŠ¤ í¬ë¡¤ë§ í•¨ìˆ˜ ì •ì˜ [í•´ë‹¹ í•¨ìˆ˜ì˜ ì •ì˜ì—ì„œ í•œê¸€ë¡œ ì…ë ¥ì„ ë°›ìœ¼ë©´ economyë‰´ìŠ¤ ì™€ ê°™ì´ ì¶œë ¥ë˜ì–´ í•´ë‹¹ ê°’ì„ GPTì—ê²Œ ìˆ˜ì • ìš”ì²­]
def print_news(section_name):
    # â˜… GPT ìˆ˜ì •ëœ ë¶€ë¶„ 1: í•œê¸€ â†’ ì˜ì–´ ì½”ë“œ ë³€í™˜
    section_code = section_dict[section_name]  # ì˜ˆ: 'ê²½ì œ' â†’ 'economy'
    # â˜… GPT ìˆ˜ì •ëœ ë¶€ë¶„ 2: URLì— ì˜ì–´ ì½”ë“œ ì‚½ì…
    url = f"https://news.daum.net/{section_code}"  # ì˜ˆ: https://news.daum.net/economy
    # â˜… GPT ìˆ˜ì •ëœ ë¶€ë¶„ 3: ì¶œë ¥ì—ëŠ” í•œê¸€ ê·¸ëŒ€ë¡œ ì‚¬ìš©
    print(f"=====> {url} {section_name} ë‰´ìŠ¤ <=====")  # ì˜ˆ: "ê²½ì œ ë‰´ìŠ¤"
    
    # ìš”ì²­ í—¤ë” ì„¤ì • (ë¸Œë¼ìš°ì €ì²˜ëŸ¼ ìš”ì²­ ë³´ë‚´ê¸° ìœ„í•´ User-Agent í¬í•¨)
    req_header = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'
    }

    # ìš”ì²­ ë³´ë‚´ê¸°
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'  # í•œê¸€ ê¹¨ì§ ë°©ì§€: ì¸ì½”ë”© ì„¤ì •ì€ soup íŒŒì‹± ì „ì—!

    # ì‘ë‹µì´ ì •ìƒì¼ ê²½ìš° HTML íŒŒì‹±
    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')

        # ë‰´ìŠ¤ ë¦¬ìŠ¤íŠ¸ í•­ëª©(li) í•˜ë‚˜ì”© ìˆœíšŒ
        for li_tag in soup.select('ul.list_newsheadline2 li'):
            a_tag = li_tag.find('a')         # ê¸°ì‚¬ ë§í¬ ì¶”ì¶œ
            if not a_tag: continue           # a íƒœê·¸ê°€ ì—†ìœ¼ë©´ ê±´ë„ˆëœ€
            link = a_tag['href']             # ë§í¬ ì£¼ì†Œ ì¶”ì¶œ
            print(link)

            # ë‰´ìŠ¤ ì œëª© ì¶”ì¶œ
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            if strong_tag:
                title = strong_tag.text.strip()  # í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ + ê³µë°± ì œê±°
                print(title)

# í•¨ìˆ˜ í˜¸ì¶œ ì˜ˆì‹œ
print_news('ê²½ì œ')  # https://news.daum.net/economy ê²½ì œ ë‰´ìŠ¤
print_news('ì‚¬íšŒ')  # https://news.daum.net/society ì‚¬íšŒ ë‰´ìŠ¤

```
